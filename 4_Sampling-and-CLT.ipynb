Let's use a Python simulation to demonstrate the Central Limit Theorem (CLT) with randomly generated patient data. For the sake of simplicity, we'll assume the distribution of patient stays follows a non-normal distribution (let's say, a gamma distribution, which can model waiting times).

First, we'll generate the population data:

```python
import numpy as np
import matplotlib.pyplot as plt

# Set a seed for reproducibility
np.random.seed(0)

# Parameters for the gamma distribution 
shape, scale = 2., 2.  

# Generate the "population"
population = np.random.gamma(shape, scale, 100000)
```

Now, we'll repeatedly sample from this population and record the sample means:


```python
# Number of samples
n_samples = 1000

# Size of each sample
sample_size = 50

# Store the sample means here
sample_means = []

for _ in range(n_samples):
    # Draw a random sample from the population
    sample = np.random.choice(population, sample_size)
    
    # Compute the sample mean
    sample_means.append(np.mean(sample))
```

Finally, we can visualize the distribution of sample means to see if it approximates a normal distribution:


According to the Central Limit Theorem, the histogram should show a distribution that looks approximately normal, centered around the true population mean, which we can compute with `np.mean(population)`. This demonstrates that even though the lengths of stay were originally gamma-distributed, the means of samples from this population are normally distributed.

Keep in mind that in real-life scenarios, you wouldn't have access to the whole "population" as we do in this simulation. Instead, you'd take a sample from the population, calculate the sample mean and use that to estimate the population mean, as described in the Central Limit Theorem.

```python
plt.hist(sample_means, bins=30, density=True)
plt.xlabel('Sample Mean Length of Stay')
plt.ylabel('Frequency')
plt.title('Distribution of Sample Means of Patient Length of Stay')
plt.show()
```
